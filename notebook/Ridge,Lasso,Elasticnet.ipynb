## ## MODEL TRAINING


```python
df= pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df['Classes'].value_counts()
```

```python
# encoding
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
## divide into indepemdent qand dependnt
x=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
x.head()
```

```python
y
```

```python
## train test splt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape
```

```python
## FEATURE SELECTION BASED ON CORR
x_train.corr()
```

```python
## check for multicolinearity
plt.figure(figsize=(12,10))
corr=x_train.corr()
sns.heatmap(corr,annot=True)
```

```python
## to check fro multicolinearity
def correlation(dataset, threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold-- domain experties
correlation(x_train,0.85) ## greater than 90%
```

```python
x_train.drop(['BUI','DC'],axis=1,inplace=True)
x_test.drop(['BUI','DC'],axis=1,inplace=True)
```

```python
x_train.shape,x_test.shape
```

```python
### feature standadization scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_sc=scaler.fit_transform(x_train)
x_test_sc=scaler.transform(x_test)
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title("before scaling")
plt.subplot(1,2,2)
sns.boxplot(data=x_train_sc)
plt.title("after scaling")
```

### LR model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

linreg = LinearRegression()
linreg.fit(x_train_sc, y_train)
y_pred = linreg.predict(x_test_sc)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("mean error", mae)
print("r2 score", score)
plt.scatter(y_test,y_pred)
```

## lasso reg

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error, r2_score

lasso = Lasso()
lasso.fit(x_train_sc, y_train)
y_pred = lasso.predict(x_test_sc)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("mean error", mae)
print("r2 score", score)
plt.scatter(y_test,y_pred)
```

## RIDGE REG

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, r2_score

ridgereg = Ridge()
ridgereg.fit(x_train_sc, y_train)
y_pred = ridgereg.predict(x_test_sc)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("mean error", mae)
print("r2 score", score)
plt.scatter(y_test,y_pred)
```

## ELASTIC NET

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error, r2_score

elnetreg = ElasticNet()
elnetreg.fit(x_train_sc, y_train)
y_pred =elnetreg.predict(x_test_sc)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("mean error", mae)
print("r2 score", score)
plt.scatter(y_test,y_pred)
```
